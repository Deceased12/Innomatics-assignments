```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print(f"Total students: {len(marks)}")
print(f"Number of pass students: {pass_count}")
print(f"Number of fail students: {fail_count}")
```

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []
for name in names:
    cleaned_names.append(name.strip().lower())

print(cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    message_length = len(message)
    print(f"Message: '{message}' | Length: {message_length}")
    if message_length > 10:
        print("  (Flagged: Message is longer than 10 characters)")
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0
for log_entry in logs:
    if log_entry == "ERROR":
        error_count += 1

print(f"Total error messages: {error_count}")
```
